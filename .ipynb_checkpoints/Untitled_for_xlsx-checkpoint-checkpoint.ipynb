### Попытка анализа количества заездов инкассаторских бригад в зависимости от дня недели и маршрута.
Данные взяты из базы по учету работы участка инкассации ПУ Участок для РОСИНКАС Читинский участок инкассации 

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
#загрузим таблицу и получем информацию о ней
#df = pd.read_csv(r'C:\Users\Галина\Desktop\DS save\incas1.csv')
df = pd.read_excel(r'C:\Users\Галина\Desktop\DS save\октябрь.xlsx')

print(df.head(2))
#df.info()
```

```python
df.columns
```

```python
#переименуем столбцы
# df.set_axis(['data','1', '2', 'nambe_trek','count_men','time','time_sum','time_night','km_trek','pount_plan',
#                         'pount_fact','3', 'mani_sum','type_trek'], axis = 'columns', inplace=True)
df.set_axis(['data', 'nambe_trek','count_men','time','time_sum','time_night','km_trek',
             'pount_plan','pount_fact', 'mani_sum','type_trek'], axis = 'columns', inplace=True)
```

```python
#выберем строки с читинским участком
#df = df.loc[df['subNameDepartment'] == '1 Читинский участок инкассации']
```

```python
print(df.head(5))
```

```python
#сформируем табл. по выбранным столбцам
df_pount = df[['data','nambe_trek','pount_fact']]
print(df_pount.head(5))
df_pount.info()
df_pount.isnull().sum()
```

```python
#удалим строки с пропусками, их мало и стат значения не несут
df_pount = df_pount.dropna() 
```

```python
df_pount.info()
```

```python
#заменим тип данных

df_pount['pount_fact'] = df_pount['pount_fact'].astype('int')
```

```python
#удалим дубликаты
print(df_pount.duplicated().sum())
df_pount = df_pount.drop_duplicates().reset_index(drop = True)
```

```python
#определим дни недели по датам
df_pount['data'] = pd.to_datetime(df_pount['data'], format='%d.%m.%Y')
df_pount['nambe_weekday'] = df_pount['data'].dt.weekday
#присвоим текстовые
#df_pount['weekday'] = pd.DatetimeIndex(df_pount)
dw = ['1_понедельник','2_вторник','3_среда','4_четверг','5_пятница','6_суббота','7_воскресенье']
def name_weekday(row):
    i = row['nambe_weekday'] 
    return dw[i]

df_pount['weekday'] = df_pount.apply(name_weekday, axis=1)


df_pount.head(5)

```

```python
#создадим словари
df_work = df_pount .loc[((df_pount['nambe_trek'] == '1') | (df_pount['nambe_trek'] == '2') | (df_pount['nambe_trek'] == '0') | 
                        (df_pount['nambe_trek'] == '3') | (df_pount['nambe_trek'] == '4') | (df_pount['nambe_trek'] == '5') |
                        (df_pount['nambe_trek'] == '6') | (df_pount['nambe_trek'] == '7') | (df_pount['nambe_trek'] == '11')|
                        (df_pount['nambe_trek'] == '12'))]#&[((df_pount['weekday'] != '6') | (df_pount['weekday'] != '7'))]
df_sub = df_pount .loc[(df_pount['nambe_trek'] == '1с') | (df_pount['nambe_trek'] == '2с') | (df_pount['nambe_trek'] == '3с') |
                       (df_pount['nambe_trek'] == '4с') | (df_pount['nambe_trek'] == '5с') |
                        (df_pount['nambe_trek'] == '6с') | (df_pount['nambe_trek'] == '7с')]
df_wos = df_pount .loc[(df_pount['nambe_trek'] == '1в') | (df_pount['nambe_trek'] == '2в') | (df_pount['nambe_trek'] == '3в') |
                       (df_pount['nambe_trek'] == '4в')]
```

```python
#получим данные по буднему дню сводная таблица
df_work_pivot = df_work.pivot_table(index = ('weekday'), columns = 'nambe_trek',
                                                values = ('pount_fact'), aggfunc = ('mean','min','max')).round(0)
df_work_pivot = df_work_pivot.fillna(0)

df_work_pivot
```

```python
sns.set(rc={'figure.figsize':(20,6)})
sns.heatmap(df_work_pivot, annot=True)
plt.show
```

```python
#отдельно mean будний день
df_work_mean_pivot = df_work.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = ('mean')).reset_index()
df_work_mean_pivot.columns=['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница','6_суббота','7_воскресенье']
df_work_mean_pivot = df_work_mean_pivot.fillna(0)
df_work_mean_pivot = df_work_mean_pivot[['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']]
df_work_mean_pivot
```

```python
col=['1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']
for i in col:
    sns.catplot(data=df_work_mean_pivot, kind='bar', x='nambe_trek', y=i)
    plt.title('заезды по будним дням недели в среднем')
```

```python
df_work_mean_obsh = df_work.pivot_table(index = 'nambe_trek',
                                                values = ('pount_fact'), aggfunc = ('mean')).reset_index()

sns.catplot(data=df_work_mean_obsh, kind='bar', x = 'nambe_trek', y = 'pount_fact')
plt.title(' Количество заездов в среднем в будний день')
plt.show()
```

```python
r = ['1','2','3','4','5','6','7','11','12','1/у','2/у','3/у','4/у','5/у','6/у','7/у','11/у','9','8','10']
#sns.set(rc={'figure.figsize':(10,5)})
for plat in r:
    e = df_pount.query('nambe_trek in @plat')
    data_p = e.pivot_table(index = ['weekday'], values = ['pount_fact'],  aggfunc =['mean'])
    sns.lineplot(data = data_p)
    plt.title(plat)
    plt.ylabel('количество точек в среднем')
    plt.show()
```

```python
df_work_mean = df_work.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = ('mean'))
sns.set(rc={'figure.figsize':(20,6)})
sns.heatmap(df_work_mean, annot=True)
plt.show
```

```python
#отдельно min будний день
df_work_min_pivot = df_work.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = 'min').reset_index()
df_work_min_pivot.columns=['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница','6_суббота','7_воскресенье']
df_work_min_pivot = df_work_min_pivot.fillna(0)
df_work_min_pivot = df_work_min_pivot[['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']]
df_work_min_pivot
```

```python
col=['1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']
for i in col:
    sns.catplot(data=df_work_min_pivot, kind='bar', x='nambe_trek', y=i)
    plt.title('заезды по будним дням недели минимальные')    
```

```python
#отдельно max будний день
df_work_max_pivot = df_work.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = 'max').reset_index()
df_work_max_pivot.columns=['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница','6_суббота','7_воскресенье']
df_work_max_pivot = df_work_max_pivot.fillna(0)
df_work_max_pivot = df_work_max_pivot[['nambe_trek','1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']]
df_work_max_pivot
```

```python
col=['1_понедельник','2_вторник','3_среда','4_четверг','5_пятница']
for i in col:
    sns.catplot(data=df_work_max_pivot, kind='bar', x='nambe_trek', y=i)
    plt.title('заезды по будним дням недели максимальные')
    plt.show()
```

```python
#получим данные за субботу
df_sub_pivot = df_sub.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = ('mean', 'min', 'max')).reset_index()
df_sub_pivot.columns = ['nambe_trek','max','mean','min']
df_sub_pivot = df_sub_pivot.fillna(0)
df_sub_pivot
```

```python
col=['max','mean','min']
for i in col:
    sns.catplot(data=df_sub_pivot, kind='bar', x='nambe_trek', y=i)
    plt.title('заезды в субботу')
```

```python
#получим данные за воскресенье
df_wos_pivot = df_wos.pivot_table(index = 'nambe_trek', columns = 'weekday',
                                                values = ('pount_fact'), aggfunc = ('mean','min', 'max')).reset_index()
#, 'min', 'max'
df_wos_pivot.columns = ['nambe_trek','max','mean','min']
df_wos_pivot
```

```python
col=['max','mean','min']

for i in col:
    sns.catplot(data=df_wos_pivot, kind='bar', x='nambe_trek', y=i)
    plt.title('заезды в воскресенье')
    
```
